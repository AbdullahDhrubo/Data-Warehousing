```python
# parsing the json data using the API for movie id 550 to 555
apnd_df = []
for i in range(6):
    response_API = requests.get(f'https://api.themoviedb.org/3/movie/55{i}?api_key=e31a75ca2478a169b405cf6ec5313235')
    df = pd.DataFrame.from_dict(response_API.json(), orient = 'index')
    result = df.reset_index().transpose()
    headers = result.iloc[0]
    new_result = pd.DataFrame(result.values[1:], columns=headers)
    apnd_df.append(new_result)
```

```python
movies_df = pd.concat(apnd_df).reset_index(drop=True)
movies_df
```

## Movies

```python
movies = movies_df[['budget', 'genres', 'id', 'imdb_id', 'original_title', 'release_date', 'revenue',
'runtime']]
movies
```

## Genres

```python
g_df = pd.DataFrame()
for genre in movies['genres']:
    for g in genre:
        g_df = g_df.append(g, ignore_index=True)
```

```python
g_df.id = g_df.id.round(0).astype(int)
genres = g_df.drop_duplicates()
genres
```

### Solving for create the genre columns which takes a column of lists and turns it into a set of columns of frequency values

```python
trans_g = g_df.transpose().reset_index(drop=True)
hdr_g = pd.DataFrame(trans_g.values[0:1], columns=trans_g.iloc[1])

hdr_g
```

```python
columns = {header: '0' for header in hdr_g}
movies = movies.assign(**columns)
```

```python
for index, row in movies.iterrows():
    for header in hdr_g:
        gen = str(row["genres"]).lower()
        if gen.__contains__(str(header).lower()):
            row[header] = "1"
        else: 
            row[header] = "0"
del movies['genres']
```

```python
genres_df = movies
genres_df
```

### Expanding out the datetime column into a table

```python
date_df = movies[['id','release_date']]
date_df
```

```python
date_df = movies[['id','release_date']]
date_df = date_df.assign(day='*', month='*', year='*', day_of_the_week='*')

for index, row in date_df.iterrows():
    date_time_obj = datetime.strptime(row["release_date"], '%Y-%m-%d')
    row['day'] = date_time_obj.day
    row['month'] = date_time_obj.month
    row['year'] = date_time_obj.year
    row['day_of_the_week'] = date_time_obj.strftime('%A')
```

## Converting runtime object

```python
movies = movies.assign(hour='*', minutes='*')

for index, row in movies.iterrows():
    runtime_obj = row["runtime"]
    hours = int(runtime_obj) // 60
    minutes = int(runtime_obj) % 60
    row['hour'] = hours
    row['minutes'] = minutes
movies
```

# SQL 

### Connection

```python
# Open a cursor to perform database operations
import sqlite3

try:
    sqliteConnection = sqlite3.connect('movies_schema.db')
    cursor = sqliteConnection.cursor()
    print("Database created and Successfully Connected to SQLite")

    sqlite_select_Query = "select sqlite_version();"
    cursor.execute(sqlite_select_Query)
    record = cursor.fetchall()
    print("SQLite Database Version is: ", record)
    cursor.close()

except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    if sqliteConnection:
        sqliteConnection.close()
        print("The SQLite connection is closed")
```

# Creating the tables in database (SQLite)

```python
#Movies Table
table_name = 'Movies'

conn = sqlite3.connect('movies_schema.db')
query = f'Create table if not Exists {table_name} (budget integer, id integer, imdb_id text, original_title text, release_date date, revenue integer, runtime integer, hour integer, minutes integer, Drama integer, Thriller integer, Comedy integer, Action integer, Adventure integer, Romance integer, Crime integer, History integer)'
conn.execute(query)
movies.to_sql(table_name,conn,if_exists='replace',index=False)
conn.commit()
conn.close()
```

```python
#Genres Table
table_name = 'Genres'

conn = sqlite3.connect('movies_schema.db')
query = f'Create table if not Exists {table_name} (id integer, name text)'
conn.execute(query)
genres.to_sql(table_name,conn,if_exists='replace',index=False)
conn.commit()
conn.close()
```

```python
#Date Table
table_name = 'Datetime'

conn = sqlite3.connect('movies_schema.db')
query = f'''Create table if not Exists {table_name} (id integer, release_date date,day integer, 
        month integer, year integer, day_of_the_week text)'''
conn.execute(query)
datetime.to_sql(table_name,conn,if_exists='replace',index=False)
conn.commit()
conn.close()
```
